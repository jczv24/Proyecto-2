### 3. Limpieza de datos

#### 3.1. Selección de variables por valores unicos

En este caso se eliminaran las variables que son irrelevantes para el modelo, como las que contienen un unico valor constante que es el asño de presentación de la prueba

```python
columns_to_drop = df.nunique()[df.nunique() == 1].index
data1 = df.drop(columns=columns_to_drop)
data1.shape
```

En este caso tambien se evalua la cantidad de registros similares que tienen algunas columnas como lo son: `estu_depto_presentacion` y `estu_depto_reside`.

```python
mask = data1['estu_depto_presentacion'] == data1['estu_depto_reside']
coincidencias = mask.mean() * 100
print(coincidencias)

nomask = data1['estu_depto_presentacion'] != data1['estu_depto_reside']
coincidencias2 = nomask.mean() * 100
print(coincidencias2)
```

Como se puede ver cerca del 98% de los registros de las dos columnas son iguales, debido a esto, se decidio conservar la columna `estu_depto_reside` ya que conceptualmente es mas relevante para predecir la situación socioeconomica

```python
data2 = data1.drop(columns=['estu_depto_presentacion'])
data2.shape
```

Este mismo fenomeno ocurre con las variables `estu_nacionalidad` y `estu_pais_reside`

```python
mask = data2['estu_nacionalidad'] == data2['estu_pais_reside']
coincidencias = mask.mean() * 100
print(coincidencias)

nomask = data2['estu_nacionalidad'] != data2['estu_pais_reside']
coincidencias2 = nomask.mean() * 100
print(coincidencias2)
```

En este caso el 99% de los registros de las dos columnas son iguales, debido a esto se decidio conservar la columna `estu_pais_reside` ya que conceptualmente es mas relevante para predecir la situación socioeconomica

```python
data3 = data2.drop(columns=['estu_nacionalidad'])
data3.shape
```

Inicialmente contamos con la información de departamento y municipio, tanto de residencia como de presentación. No obstante, se decidio conservar únicamente el departamento de residencia; por ello, las columnas de municipio de residencia y municipio de presentación se eliminarán. Además, la variabilidad municipal ya está implícita en la variable de departamento y, dado el gran número de municipios, mantener los municipios aumentaría innecesariamente la complejidad del modelo.

```python
data4 = data3.drop(columns=['estu_mcpio_presentacion','estu_mcpio_reside','cole_mcpio_ubicacion', 'cole_depto_ubicacion'])
data4.shape
```

Por otro lado, es importante observar las columnas cuyos valores unicos son muy grandes, tal es el caso de las columnas `estu_estadoinvestigacion`, `estu_estudiante`, `estu_nacionalidad`. `estu_pais_reside`. Debido a esto se asumio que todos los estudiantes residen y son de nacionalidad colombiana.

```python
data5 = data4.drop(columns=['estu_estudiante','estu_estadoinvestigacion', 'estu_pais_reside','cole_sede_principal'])
data5.shape
```

#### 3.2. Selección de variables que son identificadores

Variables que son identificadores del colegio, dadas por columnas como `estu_consecutivo`,`cole_codigo_icfes`,`cole_cod_dane_establecimiento`,`cole_cod_dane_sede`. Esto debido a que no aportan mucha información para el modelo

#### 3.3. Selección de variables por redundancia de información

Se decidio quitar las columnas `cole_cod_depto_ubicacion` , `cole_cod_mcpio_ubicacion`, `estu_cod_reside_mcpio`, `estu_cod_reside_depto` y `cole_nombre_sede` ya que se disponen de las columnas de texto: `cole_depto_ubicacion` y `cole_mcpio_ubicacion`, `estu_mcpio_reside` , `estu_depto_reside` y `cole_nombre_establecimiento`. Dichas columnas contienen la misma información por lo que se decicio remover las columnas que contienen los numeros y asi evitar redundancia en la construcción del modelo 

```python
cols_drop = [
    'cole_cod_depto_ubicacion',
    'cole_cod_mcpio_ubicacion',
    'estu_cod_reside_mcpio',
    'estu_cod_reside_depto',
    'estu_cod_mcpio_presentacion',
    'estu_cod_depto_presentacion',
    'cole_nombre_sede'
]

data7 = data6.drop(columns=cols_drop)
data7.shape
```

#### 3.4. Selección de variables por correlación

Para este caso se evaluaran las variables numericas que esten altamente correlacionadas, especialmente aquellas relacionadas con el puntaje saber 11. Para este caso al analizar el grafico de correlacion de estas variables se observa que el puntaje global tiene una gran correlación con los puntajes de las asignaturas, por lo que se puede eliminar el puntaje global y dejar los puntajes de las asignaturas.

```python
numerical_columns = data7.select_dtypes(include=[np.number]).columns
correlation_matrix = data7[numerical_columns].corr()

plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, cmap="Blues", vmin=-1, vmax=1, annot=True)
plt.title("Matriz de correlación de variables numéricas")
plt.show()
```

```python
# Eliminar puntaje global
data8 = data7.drop(columns=['punt_global'])
```

Como la variable estrato socioeconomico es la variable de interes se conservara solo las categorias pertenecientes al estrato 1 a 6 y la categoria sin estrato. Debido a que es la variable de interes en este caso no se imputaran los datos nulos, solo se eliminaran

```python
data9 = data8.dropna(subset=['fami_estratovivienda']).copy()
data9.shape
```

##### 3.5.2. Imputación datos nulos (Variables categoricas y continuas)

Como ningun porcentaje es significativamente alto, se decidio imputar los valores nulos de las variables. Pero para esto se analizara la distribución de dichas columnas para hacer la imputación de datos. Teniendo eso en cuenta para la columna `Colegio_bilingue` se decidio asiganr una variable dummy "Faltante". Como el resto de las variables presentan un porcentaje bajo las variables numericas se imputaran con la mediana y las variables categorias se imputaran con la moda. 

```python
data9['cole_bilingue'] = data9['cole_bilingue'].fillna('Faltante')
data9['cole_bilingue'].value_counts(dropna=False)
```

```python
cols = [
    'fami_tieneinternet',
    'fami_educacionpadre',
    'fami_educacionmadre',
    'fami_tieneautomovil',
    'fami_cuartoshogar',
    'fami_tienecomputador',
    'fami_tienelavadora',
    'fami_personashogar',
    'cole_caracter',
    'estu_depto_reside',               
    'estu_genero'
]

# Imputar con la moda cada columna para las variables categóricas
for col in cols:
    moda = data9[col].mode(dropna=True)[0]
    data9[col].fillna(moda, inplace=True)
```

Al analizar con detenimiento los datos, se pudo evidenciar que las columnas de `fami_educacionmadre` y `estu_genero` tenian categorias extrañas, al no tener información de dichas categorias y los pocos registros asociadaos, se decidio eliminar dichas filas.

Notamos que las columnas categoricas `estu_fechanacimiento` y `cole_nombre_establecimiento` tenian muchos valores unicos. En el caso de `estu_fechanacimiento` se decidio simplemente extrear el año de la fecha de nacimiento para no perden iformación relevante, finalmente para la columna que correponde al colegio o institución a la que fue el estudiante esta sera eliminada ya que para la variable de interes no es tan relevante
