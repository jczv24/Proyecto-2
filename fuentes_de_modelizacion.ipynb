# Fuentes de modelización

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Limpieza de los datos

### Carga de los datos después de la limpieza

```python
datos_limpios = pd.read_csv('/Users/melo/Downloads/model_data.csv')
datos_limpios
```

### Creación de las categorías del desempeño en el icfes según percentiles

```python
datos_limpios['desemp'] = pd.qcut(datos_limpios['punt_global'], q=3, labels=['Bajo', 'Medio', 'Alto'])
datos_limpios
```

### Transformación de los datos del año de nacimiento a la edad del candidato

```python
datos_limpios['estu_edad'] = 2019 - datos_limpios['estu_yearnacimiento']
datos_limpios = datos_limpios.drop(columns=['estu_yearnacimiento'])
datos_limpios
```

```python
pd.set_option('display.float_format', '{:.2f}'.format)
datos_limpios.describe()
```

### Eliminación de outliers (presentan el examen personas entre 15 y 40 años)

```python
datos_filtrados = datos_limpios[(datos_limpios['estu_edad'] >= 15) & (datos_limpios['estu_edad'] <= 40)]
datos_filtrados
```

```python
datos_filtrados.describe()
```

```python
datos_filtrados.info()
```

## Selección de variables

### Separación de variables en numéricas y categóricas (las categóricas se separan en tres grupos: cole, estu, fami)

```python
var_cat = datos_limpios.select_dtypes(include=['object']).columns
var_cat_cole = [col for col in var_cat if col.startswith('cole')]
var_cat_estu = [col for col in var_cat if col.startswith('estu')]
var_cat_estu.append('desemp_ingles')
var_cat_fami = [col for col in var_cat if col.startswith('fami')]
var_num = 'estu_edad'
target = 'punt_global'
```

### Scatter plot para ver la relación entre la edad y el puntaje global

```python
plt.scatter(datos_filtrados[var_num], datos_filtrados[target])
plt.xlabel(var_num)
```

### Boxplots entre las variables categóricas del colegio con el puntaje global

```python
for col in var_cat_cole:
    plt.figure(figsize=(10, 6))
    
    sns.boxplot(data=datos_limpios, x=col, y=target)
    
    plt.title(f'{target} por {col}')
    plt.xlabel(col)
    plt.ylabel(target)
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.tight_layout()
    
    plt.show()
```

### Boxplots entre las variables categóricas del estudiante con el puntaje global

```python
for col in var_cat_estu:
    plt.figure(figsize=(10, 6))
    
    sns.boxplot(data=datos_limpios, x=col, y=target)
    
    plt.title(f'{target} por {col}')
    plt.xlabel(col)
    plt.ylabel(target)
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.tight_layout()
    
    plt.show()
```

### Boxplots entre las variables categóricas de la familia con el puntaje global

```python
for col in var_cat_fami:
    plt.figure(figsize=(10, 6))
    
    sns.boxplot(data=datos_limpios, x=col, y=target)
    
    plt.title(f'{target} por {col}')
    plt.xlabel(col)
    plt.ylabel(target)
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.tight_layout()
    
    plt.show()
```
